```python
#counting class labels
duration_buckets = df['DURATION_BUCKET'].value_counts()
duration_buckets.axes
duration_buckets_labels = duration_buckets.axes[0].tolist()
len(duration_buckets_labels)
```

# Device Type

```python
#counting device labels
device_type = df['DEVICE_TYPE'].value_counts()
m_height = device_type.values.tolist()
device_type.axes
device_type_labels = device_type.axes[0].tolist()

device1 = []
device2 = []
device3 = []
device4 = []

for device_types in device_type_labels:
    for duration_bucket in duration_buckets_labels:
        size = len(df[(df['DEVICE_TYPE'] == device_types) & (df['DURATION_BUCKET'] == duration_bucket)].index)
        if device_types == 1:
            device1.append(size)
        elif device_types == 2:
            device2.append(size)
        elif device_types == 3:
            device3.append(size)
        else:
            device4.append(size)
```

```python
#plot preparation
import numpy as np
ind = np.arange(37)
width = 0.30
fig, ax = plt.subplots(figsize=(20,50))
device1_bars = ax.bar(ind, device1 , width, color='#ADFF2F')
device2_bars = ax.bar(ind + width, device2 , width, color='#BA70D8')
device3_bars = ax.bar(ind + 2*width, device3 , width, color='#DDFF27')
device4_bars = ax.bar(ind + 3*width, device4 , width, color='#FA70D9')

#Add some text for labels, title and axes ticks
ax.set_xlabel("Class Labels",fontsize=20)
ax.set_ylabel('Device Type Quantity',fontsize=20)
ax.set_xticks(ind) #Positioning on the x axis
ax.set_xticklabels(duration_buckets_labels,
                  fontsize = 12)
```

# MODELLING THE DATA

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
target = df['DURATION_BUCKET']
train_data = df.drop('DURATION_BUCKET',axis=1)

target.shape, train_data.shape
```

```python

clf = RandomForestClassifier(n_estimators=13)
scoring = 'accuracy'
score = cross_val_score(clf, train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# Random Forest Score
round(np.mean(score)*100, 2)
```
